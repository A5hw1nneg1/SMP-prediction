```python
import os
from PIL import Image
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
```

```python
df = pd.read_csv("/Users/harshit/Downloads/final_df.csv")
image_dir = "/Users/harshit/Downloads/"
sample_size = 5000  # Number of images to process
```

```python
def load_image(image_path):
    img = Image.open(os.path.join(image_dir, image_path))
    img = img.resize((224, 224))  # Resize image to match ResNet50 input size
    img = img.convert('RGB')  # Convert image to RGB if it's in grayscale or other formats
    return img
```

```python
base_model = ResNet50(weights="imagenet", include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False
```

```python
def extract_features(image_path):
    img = load_image(image_path)
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)  # Create batch axis
    img_array = tf.keras.applications.resnet50.preprocess_input(img_array)  # Preprocess input
    features = base_model.predict(img_array)  # Extract features
    features = features.flatten()  # Flatten features
    return features
```

```python
image_paths = df["Path"].values[:sample_size]  # Select first 10,000 images
extracted_features = [extract_features(path) for path in image_paths]
```

```python
pca_tag_vector = PCA(n_components=2)  # Specify the number of components you want to reduce to
tag_vector_pca_result = pca_tag_vector.fit_transform(tag_vector_array)
```

```python
new_df['features_PCA1'] = features_pca_result[:, 0]
new_df['features_PCA2'] = features_pca_result[:, 1]
new_df['tag_vector_PCA'] = tag_vector_pca_result.flatten()
```
